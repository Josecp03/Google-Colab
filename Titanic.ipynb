## Ejercicio 15
Mostrar por pantalla el porcentaje de menores de edad que sobrevivieron en
cada clase. En primer lugar, hay que hacerlo sin bucle y luego con un bucle
for.

### Método sin bucle

```python
# Filtrar para obtener solo menores y agrupar por clase y supervivencia
minors_survived_by_class = df[df['IsMinor'] == 'Si'].groupby(['Pclass', 'Survived']).size().unstack(fill_value=0)

# Calcular el porcentaje de supervivencia para menores en cada clase
minors_survived_by_class['Total'] = minors_survived_by_class[0] + minors_survived_by_class[1]
minors_survived_by_class['PercentageSurvived'] = (minors_survived_by_class[1] / minors_survived_by_class['Total']) * 100

# Mostrar el resultado
print("Porcentaje de menores que sobrevivieron en cada clase (sin bucle):")
display(minors_survived_by_class[['PercentageSurvived']])
```

### Método con bucle for

```python
# Get unique classes
classes = df['Pclass'].unique()

for pclass in sorted(classes):
    # Filter for minors in the current class
    minors_in_class = df[(df['IsMinor'] == 'Si') & (df['Pclass'] == pclass)]

    # Count total minors and surviving minors in the class
    total_minors_in_class = len(minors_in_class)
    surviving_minors_in_class = len(minors_in_class[minors_in_class['Survived'] == 1])

    # Calculate percentage
    if total_minors_in_class > 0:
        percentage_survived = (surviving_minors_in_class / total_minors_in_class) * 100
        print(f"Clase {pclass}: {percentage_survived:.2f}%")
    else:
        print(f"Clase {pclass}: No hay menores en esta clase.")
```
